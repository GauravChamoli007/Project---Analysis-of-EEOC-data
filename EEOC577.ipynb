```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv("Education vs Per Capita Personal Income by County Cleaned V2.csv")
```

```python
df.drop('County_FIPS', axis=1, inplace=True)
```

```python
df
```

```python
col = df.columns
col = col[:-1]
```

```python
# Create histograms for each column
for column in col:
    plt.figure(figsize=(6, 4))
    plt.title(f'Histogram of {column}')
    sns.histplot(df[column], kde=True)
    plt.show()
```

```python
# Create box plots for numeric columns
numeric_columns = df.select_dtypes(include=[np.number]).columns
for column in numeric_columns:
    plt.figure(figsize=(6, 4))
    plt.title(f'Box Plot of {column}')
    sns.boxplot(data=df[column])
    plt.show()
```

```python
# Create clustering charts for numeric columns
for column in numeric_columns:
    X = df[[column]]
    kmeans = KMeans(n_clusters=3)  # Adjust the number of clusters as needed
    df['Cluster'] = kmeans.fit_predict(X)

    plt.figure(figsize=(6, 4))
    plt.title(f'Clustering of {column}')
    sns.scatterplot(data=df, x=column, y='Cluster', hue='Cluster', palette='viridis')
    plt.xlabel(column)
    plt.ylabel('Cluster')
    plt.show()
```

```python
features = df[['Per_capita_personal_income_2019', 'Per_Capita_Personal_Income_2020','Associate_Degree_Percentage_2016_2020','Bachelor_Degree_Percentage_2015_2019']]
target = df['Per_Capita_Personal_Income_2021']
```

```python
from sklearn.linear_model import LinearRegression

# Create a linear regression model
model = LinearRegression()

# Fit the model on your data
model.fit(features, target)
```

```python
indicators_2022 = pd.DataFrame({
    'Per_capita_personal_income_2019': df['Per_Capita_Personal_Income_2020'],
    'Per_Capita_Personal_Income_2020': df['Per_Capita_Personal_Income_2021'],
    'Associate_Degree_Percentage_2016_2020':df['Associate_Degree_Percentage_2016_2020'] ,
    'Bachelor_Degree_Percentage_2015_2019': df['Bachelor_Degree_Percentage_2015_2019']
})
```

```python
per_capita_income_2022 = model.predict(indicators_2022)
per_capita_income_2022
```

```python
import pandas as pd
import folium

# Load the merged dataset
merged_data = df_new.copy()  # Replace with your dataset filename

# Calculate the central coordinates for the map
latitude_center = merged_data['LAT'].mean()
longitude_center = merged_data['LONG'].mean()

# Create a basic map
m = folium.Map(location=[latitude_center, longitude_center], zoom_start=4)  # Set the initial map location and zoom level

# Add markers for colleges and economic indicators
for index, row in merged_data.iterrows():
    college_name = row['Name']
    college_lat = row['LAT']
    college_long = row['LONG']
    per_capita_income = row['Per_Capita_Personal_Income_2021']

    # Create a marker for each college with a popup displaying relevant information
    folium.Marker(
        location=[college_lat, college_long],
        popup=f"{college_name}<br>Per Capita Income: ${per_capita_income}",
    ).add_to(m)

# Save the map to an HTML file or display it
m.save('colleges_and_income_map.html')
```
